```python
%%capture
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import deque
import matplotlib
import yasa
from scipy.signal import welch

import mne

import pyeeg
%matplotlib qt
```

## Declare band types range

```python
bands=[(0.5, 4, 'Delta'), (4, 8, 'Theta'), (8, 14, 'Alpha'), (14, 30, 'Beta'), (16, 45, 'Gamma')]
```

## Extract bandpower

```python
def PSD_calculation_on_each_epoch(epochs):
    data = epochs.get_data()
    sf = epochs.info['sfreq']
    win = int(4 * sf)  # Window size is set to 4 seconds
    freqs, psd = welch(data, sf, nperseg=win, axis=-1)
    return freqs, psd
```

```python
fileTypeList = ["_lo", "_hi"]

for iterator in range (0, 2):
    fileType = fileTypeList[iterator]
    for x in range(1, 49):
        fileName = str(x) + fileType
        print(fileName)
        if (x < 10):
            path ="D:\\New folder\\Semester 7\\CAT 400\\STEW New Epochs\\0" + fileName + "_epo.fif"
        elif (x >=10):
            path ="D:\\New folder\\Semester 7\\CAT 400\\STEW New Epochs\\" + fileName + "_epo.fif"
        else:
            continue

        print("sub" + str(x) + ":")
        
        epochs = mne.read_epochs(path, proj=True, preload=True, verbose=None)
#         print(epochs.get_data().shape)
        # epoch.plot(n_channels=14, scalings='auto')
        freqs, psd = PSD_calculation_on_each_epoch(epochs)
        
        # Calculate the bandpower on 3-D PSD array
        bandpower = yasa.bandpower_from_psd_ndarray(psd, freqs, bands)
        number_of_epochs = bandpower.shape[1]
        
        if x < 10:
            destination_folder = "D:\\New folder\\Semester 7\\CAT 400\\STEW Features\\0" + fileName + ".csv"
        else:
            destination_folder = "D:\\New folder\\Semester 7\\CAT 400\\STEW Features\\" + fileName + ".csv"

        pd.DataFrame(bandpower.transpose(1,2,0).reshape(number_of_epochs, -1)).to_csv(destination_folder, 
           header=['AF3_Delta', 'AF3_Theta', 'AF3_Alpha', 'AF3_Beta','AF3_Gamma', 
                   'F7_Delta', 'F7_Theta', 'F7_Alpha', 'F7_Beta', 'F7_Gamma',
                   'F3_Delta', 'F3_Theta', 'F3_Alpha', 'F3_Beta', 'F3_Gamma', 
                   'FC5_Delta', 'FC5_Theta', 'FC5_Alpha', 'FC5_Beta', 'FC5_Gamma',
                   'T7_Delta', 'T7_Theta', 'T7_Alpha', 'T7_Beta', 'T7_Gamma',
                   'P7_Delta', 'P7_Theta', 'P7_Alpha', 'P7_Beta', 'P7_Gamma', 
                   'O1_Delta', 'O1_Theta', 'O1_Alpha', 'O1_Beta', 'O1_Gamma', 
                   'O2_Delta', 'O2_Theta', 'O2_Alpha', 'O2_Beta', 'O2_Gamma', 
                   'P8_Delta', 'P8_Theta', 'P8_Alpha', 'P8_Beta', 'P8_Gamma',
                   'T8_Delta', 'T8_Theta', 'T8_Alpha', 'T8_Beta', 'T8_Gamma',
                   'FC6_Delta', 'FC6_Theta', 'FC6_Alpha', 'FC6_Beta', 'FC6_Gamma', 
                   'F4_Delta', 'F4_Theta', 'F4_Alpha', 'F4_Beta', 'F4_Gamma', 
                   'F8_Delta', 'F8_Theta', 'F8_Alpha', 'F8_Beta', 'F8_Gamma',
                   'AF4_Delta', 'AF4_Theta', 'AF4_Alpha', 'AF4_Beta', 'AF4_Gamma'])
```
